```python
import pandas as pd
import numpy as np
```

```python
file = "iris.csv"
iris_df = pd.read_csv(file)
iris_df.head()
```

```python
iris_df.info()
```

```python
iris_df.groupby(iris_df['class']).count()
```

```python
# drop the non-numeric column i.e. class
new_iris_df = iris_df.drop(['class'], axis=1)
new_iris_df.head()
```

```python
# reorder columns
new_iris_df = new_iris_df[['sepal_length','petal_length','sepal_width','petal_width']]
new_iris_df.head()
```

```python
# save output
output = "new_iris_data.csv"
new_iris_df.to_csv(output, index=False)
```
