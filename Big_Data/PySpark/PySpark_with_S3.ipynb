```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("CloudETL").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
# Read in data from S3 Buckets
from pyspark import SparkFiles
url ="https://notaleakybucket.s3.us-east-2.amazonaws.com/user_data.csv"
spark.sparkContext.addFile(url)
user_data_df = spark.read.csv(SparkFiles.get("user_data.csv"), sep=",", header=True, inferSchema=True)
```

```python
user_data_df.show()
```

```python
# read in user payments csv
url ="https://notaleakybucket.s3.us-east-2.amazonaws.com/user_payment.csv"
spark.sparkContext.addFile(url)
user_payment_df = spark.read.csv(SparkFiles.get("user_payment.csv"), sep=",", header=True, inferSchema=True)

# Show DataFrame
user_payment_df.show()
```

```python
# Join the tables using inner join (joins on common column, which is "username")
data_payment_df = user_data_df.join(user_payment_df, on="username", how="inner")
data_payment_df.show()
```

```python
all_data_df = data_payment_df.dropna()
all_data_df.show()
```

```python
# import dependency to allow column manipulation
from pyspark.sql.functions import col

# filter for columns with active users
filtered_df = all_data_df.filter(col("active_user") == True)
filtered_df.show()
```

```python
# create dataframe to match SQL table for active_user
active_user_df = filtered_df.select(["id","first_name","last_name","username"])
active_user_df.show()
```

```python
# create dataframe to match SQL table of billing_info
billing_info_df = filtered_df.select(["billing_id","street_address","state","username"])
billing_info_df.show()
```

```python
# create payment_info dataframe to match SQL table
payment_info_df = filtered_df.select(["billing_id","cc_encrypted"])
payment_info_df.show()
```
